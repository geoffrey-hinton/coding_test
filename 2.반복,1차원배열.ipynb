```python
# 10810 번
# 바구니 N개, 1-N까지 적힌 공 N개
# M번 공을 넣을때 각 바구니에 어떤 공이 들어있는지 구하셈
# 공을 넣을 바구니는 연속, 공이 이미 있는 경우에는 들어있는거 빼고 새로운공 투입

# 첫째 줄에 N과 M주어짐
n, m = map(int, input().split(" "))

# 둘째 줄부터 M번째 줄까지 i, j, k 가 주어짐

# i번째부터 j번째까지 k번 공을 넣는다는 뜻

# 공이 들어있지 않다면 0 출력

temp_li = [0] * n
for _ in range(m):
    i, j, k = map(int, input().split(" "))
    temp_num = i
    for temp_num in range(i-1, j, 1):
        temp_li[temp_num] = k
        
print(*temp_li)
    
```

```python
# 10813번 

# 바구니의 개수 N, 공을 바꾸는 횟수 M
# 바구니의 번호와 공의 번호는 같음 공은 하나씩
# 첫째줄 N, M
# 이후로 공을 교환할 방법이 주어짐

n, m = map(int, input().split(" "))
basket = [i for i in range(1, n+1)]

for _ in range(m):
    i, j = map(int, input().split(" "))
    temp_num = basket[i-1]
    basket[i-1] = basket[j-1]
    basket[j-1] = temp_num
print(*basket)
```
