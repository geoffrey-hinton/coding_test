```python
# 25304 번

# line 1 : 총 금액 X
# line 2 : 물건 종류
# 이후 : 물건의 가격, 개수
# 총금액이 영수증과 같으면 Yes or No

tt = int(input())
iter = int(input())
total = 0
for i in range(iter):
    a, b = map(int, input().split(" "))
    total += a * b
if tt == total:
    print("Yes")
else:
    print("No")
```

```python
# 2439번

# 숫자 n 입력받고 오른쪽 정렬해서 별 출력

n = int(input(""))

for i in range(1, n+1):
    temp_line = " " * (n-i)
    temp_star = i * "*"
    print(temp_line + temp_star)
```
