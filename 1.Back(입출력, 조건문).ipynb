```python
# 2480번

# 주사위관련
# 3개 다 같으면 10,000 + (같은 눈) * 1000
# 2개 1,000 + (같은 눈) * 100
# 다 다르면 가장 큰 눈 * 100
```

```python
a, b, c = map(int, input().split(" "))

temp_li = [a, b, c]
same_num = 0
for i in temp_li:
    temp_count = temp_li.count(i)
    if temp_count == 3:
        print(10000 + i * 1000)
        same_num += 1
        break
    elif temp_count == 2:
        print(1000 + i * 100)
        same_num += 1
        break
    else:
        pass
if same_num == 0:
    print(max(temp_li) * 100)
```
